````python
library = {}

def add_book(book_name):
    library[book_name] = True

def borrow_book(book_name):
    if library.get(book_name):
        library[book_name] = False
        print(f"You borrowed '{book_name}'")
    else:
        print(f"'{book_name}' is not available")

def return_book(book_name):
    if book_name in library:
        library[book_name] = True
        print(f"You returned '{book_name}'")
    else:
        print(f"'{book_name}' does not belong to this library")

def print_available_books():
    print("\nAvailable books:")
    for book, available in library.items():
        if available:
            print(book)

# Example usage
add_book("Python Basics")
add_book("Data Science 101")
add_book("Machine Learning")

borrow_book("Python Basics")
return_book("Python Basics")
print_available_books()
````

# ðŸ“˜ NumPy Arrays Basics

## 1. Array Creation
**Notes:**
- Arrays are created using `np.array()` or helper functions like `np.arange()`, `np.zeros()`, `np.ones()`, `np.full()`, `np.random.rand()`.
- Arrays can be 1D, 2D, or higher dimensions.

```python
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D array:", arr1)

# 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array:\n", arr2)

# Range of numbers
arr3 = np.arange(1, 11)  # numbers 1 to 10
print("Range array:", arr3)

# Zeros and ones
zeros = np.zeros((2, 3))
ones = np.ones((2, 3))
print("Zeros:\n", zeros)
print("Ones:\n", ones)

# Random numbers
rand_arr = np.random.rand(3, 3)
print("Random array:\n", rand_arr)
```

---

## 2. Indexing
**Notes:**
- Access elements using square brackets `[]`.
- Works like Python lists but supports multi-dimensional indexing.

```python
arr = np.array([10, 20, 30, 40, 50])
print("First element:", arr[0])
print("Last element:", arr[-1])

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Element at row 0, col 1:", matrix[0, 1])  # 2
print("Element at row 1, col 2:", matrix[1, 2])  # 6
```

---

## 3. Slicing
**Notes:**
- Extract subarrays using `start:end:step`.
- Works in multiple dimensions.

```python
arr = np.arange(1, 11)  # [1,2,...,10]
print("Slice 2 to 6:", arr[2:7])       # [3,4,5,6,7]
print("Every second element:", arr[::2])  # [1,3,5,7,9]

matrix = np.arange(1, 10).reshape(3, 3)
print("Matrix:\n", matrix)
print("First row:", matrix[0, :])
print("First column:", matrix[:, 0])
print("Submatrix:\n", matrix[0:2, 1:3])
```

---

## 4. Operations
**Notes:**
- Arrays support element-wise arithmetic and broadcasting.
- Common operations: `+`, `-`, `*`, `/`, `**`.
- Aggregate functions: `.sum()`, `.mean()`, `.max()`, `.min()`.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Element-wise operations
print("Addition:", arr1 + arr2)
print("Multiplication:", arr1 * arr2)

# Scalar operations
print("Multiply by 10:", arr1 * 10)

# Aggregate functions
print("Sum:", arr1.sum())
print("Mean:", arr1.mean())
print("Max:", arr1.max())
print("Min:", arr1.min())
```
